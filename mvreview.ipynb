```python

data_positive=pd.read_excel("/content/drive/MyDrive/Dataset/movie review/positive.xlsx")
data_negative=pd.read_excel("/content/drive/MyDrive/Dataset/movie review/negative.xlsx")
```

```python
data_positive["class"]=1
data_negative["class"]=0
```

```python
data_positive_manual_testing = data_positive.tail(10)
for i in range(21416,21406,-1):
    data_positive.drop([i],axis=0, inplace=True)#removing those 10 values from original dataset

data_negative_manual_testing = data_negative.tail(10)
for i in range(21416,21406,-1):
    data_negative.drop([i],axis=0, inplace=True)#removing those 10 values from original dataset
```

```python
data_manual_testing = pd.concat([data_negative_manual_testing, data_positive_manual_testing], axis=0)
data_manual_testing.to_csv("manual_testing.csv ")
```

```python
data_merge = pd.concat([data_negative,data_positive],axis=0)
data_merge.head(10)#displaying first 10 values after above operation
```

```python
print(data_merge.groupby(['review'])['sentiment'].count())
data_merge.groupby(['review'])['sentiment'].count().plot(kind="bar")
plt.title("Movie review",size=20)
plt.xlabel("Category",size=20)
plt.ylabel(" count",size=20)
plt.show()
```

```python
print(data_merge.groupby(['class'])['sentiment'].count())
print("0 = negative news\n1 = positive news")
data_merge.groupby(['class'])['sentiment'].count().plot(kind="pie")
plt.title("Movie review",size=20)
plt.show()
```
