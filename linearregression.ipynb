```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
t = np.array([5,7,12,16,20]).reshape(-1,1)
m = np.array([40,120,180,210,240])#target vector
```

```python
LR = LinearRegression()
LR.fit(t,m)
```

```python
plt.scatter(t,m,color='black')
y_pred = LR.predict(t)
print(y_pred)
plt.plot(t,y_pred, color='blue',linewidth=3,marker="H")
plt.xlabel('Time')
plt.ylabel('Mass')
plt.title('Linear Regression')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
distances = np.array([1,2,5,10,15,20,21,22,24,25,26,27,28,29,30,35,40,41,47,50]).reshape(-1,1)
probabilities = np.array([1,1,1,1,1,1,0.9,0.85,0.73,0.67,0.5,0.47,0.39,0.31,0.25,0.15,0,0,0,0])
threshold = 0.
binary_labels = (probabilities > threshold)
logr = LogisticRegression()
logr.fit(distances,binary_labels)
```

```python
p = logr.predict([[25]])
print(p)
```

```python
if p==[True]:
 print("goal")
else:
  print("nogoal")
```

```python
dist = np.linspace(1,50,100).reshape(-1,1)
print(dist)
prob = logr.predict_proba(dist)
print(prob)
```

```python
import matplotlib.pyplot as plt
plt.scatter(distances, binary_labels, color='black',label='Data')
plt.plot(dist, prob,color='blue',label='Logistict Regression')
plt.title('Distance vs probabilityof scoring a goal')
plt.xlabel('Distance')
plt.ylabel('probability')
plt.legend()
plt.grid(False)
plt.show()
```
